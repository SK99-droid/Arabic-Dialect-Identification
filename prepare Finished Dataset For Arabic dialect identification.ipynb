```python
import pandas as pd

import glob

path = r'C:\Users\Salma\Downloads\Arabic_Preprocissing_data-main\Arabic_Preprocissing_data-main\RenameDataset' # use your path

all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:

    df = pd.read_csv(filename, index_col=None, header=0)

    li.append(df)

data = pd.concat(li, axis=0, ignore_index=True)
```
